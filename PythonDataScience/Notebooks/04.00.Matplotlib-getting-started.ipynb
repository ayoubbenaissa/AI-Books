```python
import numpy as np
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
#MATLAB style interface:
#working with pyplot is similar to working with MATLAB: 

plt.figure() # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
#this is equivalent but with two columns rather than 2 rows:
plt.figure()

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
```

```python
#working with 2*2 subplots:

plt.figure()

plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title('Sin function:')

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title('Cos function')

plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x))
plt.title('Exp function')

plt.subplot(2, 2, 4)
plt.plot(x, np.log(2+x))
plt.title('modified Log function')

#plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,wspace=0.35)
plt.subplots_adjust(hspace=0.5)
```

```python
#Object-Oriented Interface:

# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(1, 2)
# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2, 2)
# Call plot() method on the appropriate object

ax[0, 0].plot(x, np.sin(x))
ax[0, 0].set_title('Sin')
ax[0, 1].plot(x, np.cos(x))
ax[0, 1].set_title('Cos')
ax[1, 0].plot(x, np.exp(x))
ax[1, 0].set_title('Exp')
ax[1, 1].plot(x, np.log(2+x))
ax[1, 1].set_title('modif Log')

plt.subplots_adjust(hspace=0.5)
```
