```python
testset.groupby('year').sum()
```

```python
testset.groupby('year')['number'].sum()
```

```python
planets.groupby('method')['orbital_period'].median()
```

```python
display(testset.groupby('method')['year'])
```

```python
planets.groupby('method')['year'].describe().unstack() 
#perform a set of aggregations that describe each group in the data
```

```python
planets.groupby('method')['year'].describe()
```

```python
decade = 10 * (planets['year'] // 10)
decade = decade.astype(str) + 's'
decade.name = 'decade'
planets.groupby(['method', decade])['number'].sum().unstack().fillna(0)

#this will count discovered planets by method and by decade
```
